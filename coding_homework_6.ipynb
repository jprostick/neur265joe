```python
#importing packages

import matplotlib.pyplot as plt
import numpy as np
```

```python
#load data
acc_fa = np.array(np.loadtxt("https://raw.githubusercontent.com/jprostick/neur265joe/main/acc_fa_hw6.csv", delimiter = ","))
acc_hits = np.array(np.loadtxt("https://raw.githubusercontent.com/jprostick/neur265joe/main/acc_hits_hw6.csv", delimiter = ","))
lc_fa = np.array(np.loadtxt("https://raw.githubusercontent.com/jprostick/neur265joe/main/lc_fa_hw6.csv", delimiter = ","))
lc_hits = np.array(np.loadtxt("https://raw.githubusercontent.com/jprostick/neur265joe/main/lc_hits_hw6.csv", delimiter = ","))
```

```python
#print shapes of each array
print(np.shape(acc_fa))
print(np.shape(acc_hits))
print(np.shape(lc_fa))
print(np.shape(lc_hits))
```

```python
# create time variable
row, col = np.shape(acc_fa)
time = np.linspace(0,4, col)
```

```python
#determine sampling frequency
hz = col/time[-1]
hz
```

```python
#plot acc and lc hits on same plot, plot acc and lc false alarms on second plot

fig, (ax1,ax2) = plt.subplots(2, 1)
ax1.plot(acc_hits[0], color = "red")
ax1.plot(lc_hits[0], color = "blue")
ax2.plot(acc_fa[0], color = "red")
ax2.plot(lc_fa[0], color = "blue")



ax2.set_xlabel('Time')
ax2.set_ylabel('Voltage')
ax1.set_ylabel('Voltage')



```

For acc, I see about 5 hz. It's difficult to pick a domanant frequency, but for lc I see about 7 hz. The 7 hz oscillation in the lc is not very dominant, whereas the 5 hz frequency in the acc is more dominant and easy to see. I do not see a consistent relationship between the brain areas by visual inspection, but I do see several spikes in lc data which seem to correlate to peaks/troughs in the acc. This leads me to believe that these areas of the brain undergo some communication during these trials.

```python
# computing coherence
dt = time[1] - time[0] 
N = acc_hits.shape[1]
# Compute the Fourier transforms
xf = np.array([np.fft.rfft(x - x.mean()) for x in acc_hits])  
yf = np.array([np.fft.rfft(y - y.mean()) for y in lc_hits])  

# Compute the spectra
T = time[-1]
scale = 2 * dt**2 / T
Sxx = scale * (xf * xf.conj()).mean(0)  
Syy = scale * (yf * yf.conj()).mean(0)  
Sxy = scale * (xf * yf.conj()).mean(0) 

# Compute the coherence.
cohr = np.abs(Sxy) / (np.sqrt(Sxx) * np.sqrt(Syy))

f = np.fft.rfftfreq(N, dt)                     
plt.plot(f, cohr.real)                      
plt.xlim([0, 50])                           
plt.ylim([0, 1])                            
plt.xlabel('Frequency [Hz]')                
plt.ylabel('Coherence')
plt.title('Coherence between acc_hits and lc_hits')
```

Coherence between the data seems to be highest around either 2 hz or 7 hz.

```python
#plotting hit coherence and false alarm coherence on same plot.

xa = np.array([np.fft.rfft(x - x.mean()) for x in acc_fa])  
ya = np.array([np.fft.rfft(y - y.mean()) for y in lc_fa])  

# Compute the spectra

Sxx2 = scale * (xa * xa.conj()).mean(0)  
Syy2 = scale * (ya * ya.conj()).mean(0)  
Sxy2 = scale * (xa * ya.conj()).mean(0) 

# Compute the coherence.
cohr2 = np.abs(Sxy2) / (np.sqrt(Sxx2) * np.sqrt(Syy2))

f = np.fft.rfftfreq(N, dt)                     
plt.plot(f, cohr.real, color = "red")      
plt.plot(f, cohr2.real, color = "blue")                 
plt.xlim([0, 50])                           
plt.ylim([0, 1])                            
plt.xlabel('Frequency [Hz]')                
plt.ylabel('Coherence')
plt.title('Coherence between acc_hits and lc_hits AND Coherence between acc_fa and lc_fa')
```

Coherence between false alarms seem highest at around 2 hz, whereas for hits coherence seems highest at heither 2 or 7 hz. In general, coherence is higher between hit data than between false alarm data. It seems that there is greater communication between the acc and lc when there is a hit, as opposed to when there is a false alarm.
