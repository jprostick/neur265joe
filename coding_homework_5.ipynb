```python
#importing packages

import matplotlib.pyplot as plt
import numpy as np
```

```python
#importing data

acc_hits = np.loadtxt("https://raw.githubusercontent.com/jprostick/neur265joe/main/acc_hits_hw5.csv", delimiter = ",")
acc_fa = np.loadtxt("https://raw.githubusercontent.com/jprostick/neur265joe/main/acc_fa_hw5.csv", delimiter = ",")
```

```python
#printing shapes of acc_hits and acc_fa

print(np.shape(acc_hits))
print(np.shape(acc_fa))
```

```python
#creating time variable w/ same number of values as samples in each trial

time = np.linspace(0,4,8001)
```

```python
#calculate sampling frequency

row, col = np.shape(acc_hits)
hz = col/time[-1]
hz
```

```python
#plotting first trials of acc_hits and acc_fa data sets

fig, (ax1,ax2) = plt.subplots(2, 1)
ax1.plot(acc_hits[0])
ax2.plot(acc_fa[0])

ax1.set_title('Hits')
ax1.set_ylabel('Voltage')

ax2.set_title('False Alarm')
ax2.set_xlabel('Time')
ax2.set_ylabel('Voltage')

```

For acc_hits, dominant frequency looks like it is approximately 5 Hz. Dominant frequency for acc_fa looks to be around 2-3 Hz. It appears that acc_hits has a slightly higher donimant freqency.

```python

#plotting ERP for acc_hits

fig, ax3 = plt.subplots()
ax3.plot(time, acc_hits.mean(0))

ax3.hlines(0, 0,time[-1], color = "black")
ax3.vlines(2,-50,50, color = "black")

ax3.set_xlabel('Time (s)')
ax3.set_ylabel('Voltage (uV)')
ax3.set_title('Event-Related Potential (ERP) for acc_hits')
```

Yes. There is a clear spike in LFP activity following the pressing of the screen.

```python
#plotting ERPs for acc_hits & acc_fa on the same plot
fig, ax4 = plt.subplots()
ax4.plot(time, acc_hits.mean(0))
ax4.plot(time, acc_fa.mean(0))

ax4.hlines(0, 0,time[-1], color = "black")
ax4.vlines(2,-50,50, color = "black")

ax4.set_xlabel('Time (s)')
ax4.set_ylabel('Voltage (uV)')
ax4.set_title('Event-Related Potential (ERP) for acc_hits and acc_fa')
```

Yes. There does not seem to be a spike in LFP activity surrounding a false alarm, whereas there is a clear spike in activity following a hit.
