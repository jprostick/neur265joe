```python
#importing packages
import numpy as np
import matplotlib.pyplot as plt
```

```python
#importing .csvs into arrays
spikes_left_trial1 = np.array(np.loadtxt("https://raw.githubusercontent.com/jprostick/neur265joe/main/spikes_left_trial1.csv", delimiter = ","))
spikes_right_trial1 = np.array(np.loadtxt("https://raw.githubusercontent.com/jprostick/neur265joe/main/spikes_right_trial1.csv", delimiter = ","))
spikes_left_trial2 = np.array(np.loadtxt("https://raw.githubusercontent.com/jprostick/neur265joe/main/spikes_left_trial2.csv", delimiter = ","))
spikes_right_trial2 = np.array(np.loadtxt("https://raw.githubusercontent.com/jprostick/neur265joe/main/spikes_right_trial2.csv", delimiter = ","))
```

```python
#printing sizes of data arrays
print(np.size(spikes_left_trial1))
print(np.size(spikes_right_trial1))
print(np.size(spikes_left_trial2))
print(np.size(spikes_right_trial2))
```

```python
#calculating and printing mean firing frequencies
hzL1 = np.size(spikes_left_trial1)/30
hzR1 = np.size(spikes_right_trial1)/30
hzL2 = np.size(spikes_left_trial2)/30
hzR2 = np.size(spikes_right_trial2)/30
print(hzL1)
print(hzR1)
print(hzL2)
print(hzR2)
```

There does not seem to be a higher firing rate in the neuron depending on which direction it is about to turn. In the first trial, the neuron fires at a higher freq before a left turn. However, in the second trial, the neuron fires at a higher freq before a right turn.

```python
#creating subpplots
fig, (ax1,ax2) = plt.subplots(2, 1)

#plotting spike train data
ax1.scatter(spikes_left_trial1,spikes_left_trial1[:]/spikes_left_trial1[:],color = "red",s=3)
ax1.scatter(spikes_right_trial1,2*spikes_right_trial1[:]/spikes_right_trial1[:],s=3)
ax2.scatter(spikes_left_trial2,spikes_left_trial2[:]/spikes_left_trial2[:],color = "red", s=3)
ax2.scatter(spikes_right_trial2,2*spikes_right_trial2[:]/spikes_right_trial2[:], s=3)

#plotting stuff
ax1.set_xlabel("Time (s)")
ax2.set_xlabel("Time (s)")
ax1.set_ylabel("Trial 1 L/R Signal fires")
ax2.set_ylabel("Trial 2 L/R Signal fires")
```

The neuron seems to be fairly bursty, before either a right or left turn, in each trial. There is not a consistent firing pattern, activity seems bursty/irregular.

```python
#creating ISI arrays
sL1ISI = np.diff(spikes_left_trial1)
sR1ISI = np.diff(spikes_right_trial1)
sL2ISI = np.diff(spikes_left_trial2)
sR2ISI = np.diff(spikes_right_trial2)
```

```python
#creating subplots @ bins
fig,(ax2,ax3) = plt.subplots(2, 1)
fig3,(ax4,ax5) = plt.subplots(2, 1)
bins = np.linspace(0, 2, 200)

#binning ISIs
sL1ISIbin = np.digitize(sL1ISI, bins)
sR1ISIbin = np.digitize(sR1ISI, bins)
sL2ISIbin = np.digitize(sL2ISI, bins)
sR2ISIbin = np.digitize(sR2ISI, bins)

#making histogram of bins of ISIs
ax2.hist(sL1ISIbin, color = "red")
ax3.hist(sR1ISIbin)
ax4.hist(sL2ISIbin,color = "red")
ax5.hist(sR2ISIbin)

#plotting stuff
ax2.set_title('Binned sLeft1 ISI')
ax3.set_title('Binned sRight1 ISI')
ax4.set_title('Binned sLeft2 ISI')
ax5.set_title('Binned sRight2 ISI')

ax2.set_ylabel("frequency")
ax3.set_ylabel("frequency")
ax4.set_ylabel("frequency")
ax5.set_ylabel("frequency")

ax2.set_ylim(0,50)
ax3.set_ylim(0,50)
ax4.set_ylim(0,50)
ax5.set_ylim(0,50)


ax5.set_xlabel("ISI (s)")
```

This cell does seem to fire in bursts. Short ISIs occur in greatest frequency, but there are also sporadic longer ISIs in the cell, before either a right or left turn. It seems there is less consistency between ISIs of trials 1 & 2 before a right turn, so perhaps the firing is less regular and more bursty in general before a right turn. Overall, both an upcoming left turn and an upcoming right turn seem to induce sporadic, bursty activity in the cell.

In a new Code Cell, calculate the firing rate across time for each of your four arrays (trial 1 left-turn, trial 1 right-turn, trial 2 left-turn, and trial 2 right-turn).
```python
#creating bins & subplots
bins2 = np.linspace(0, 30, 30)
fig4, (ax6,ax7) = plt.subplots(2, 1)

#binning L/R trial data
sL1Hzbin = np.digitize(spikes_left_trial1,bins2)
sR1Hzbin = np.digitize(spikes_right_trial1,bins2)
sL2Hzbin = np.digitize(spikes_left_trial2,bins2)
sR2Hzbin = np.digitize(spikes_right_trial2,bins2)

#creating firing rate arrays
freqsL1 = np.linspace(0,30,30)
freqsR1 = np.linspace(0,30,30)
freqsL2 = np.linspace(0,30,30)
freqsR2 = np.linspace(0,30,30)

#setting all values to 0
for a in range(30):
  freqsL1[a]=0
  freqsR1[a]=0
  freqsL2[a]=0
  freqsR2[a]=0

#how many actions potentials fired in each second of each trial?
for i in range(np.size(sL1Hzbin)):
  freqsL1[sL1Hzbin[i]] = freqsL1[sL1Hzbin[i]]+1
for i in range(np.size(sR1Hzbin)):
  freqsR1[sR1Hzbin[i]] = freqsR1[sR1Hzbin[i]]+1

for i in range(np.size(sL2Hzbin)):
  freqsL2[sL2Hzbin[i]] = freqsL2[sL2Hzbin[i]]+1
for i in range(np.size(sR2Hzbin)):
  freqsR2[sR2Hzbin[i]] = freqsR2[sR2Hzbin[i]]+1

#plotting
ax6.plot(bins2,freqsL1, color = "red")
ax6.plot(bins2,freqsR1, color = "blue")

ax7.plot(bins2,freqsL2, color = "red")
ax7.plot(bins2,freqsR2, color = "blue")

ax6.set_ylabel("Firing rate")
ax7.set_ylabel("Firing rate")
ax7.set_xlabel("Time bins (s)")
```
